```python
import math

RADIUS = 5

def rad(x, y, z):
    return math.sqrt(x**2 + y**2 + z**2)

for x in range(-RADIUS, RADIUS+1):
    for y in range(-RADIUS, RADIUS+1):
        for z in range(-RADIUS, RADIUS+1):
            r = rad(x, y, z)
            if r < RADIUS:
                rand = (RADIUS - int(r)) * 10
                print(f"execute if predicate ao:random_{rand} positioned ~{x} ~{y} ~{z} if block ~ ~ ~ #ao:plantlife run setblock ~ ~ ~ minecraft:air destroy")
```
